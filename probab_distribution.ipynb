```python
import numpy as np
from numpy import random
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import  math 
```

```python
#  plotting stype
sns.set_style("white")
sns.set_context("paper")
np.random.seed(42)
```

```python
def plot(X_data,pdf_data,title=""):
  "this is going to show the diff  between the pdf and the data"
  "X_data: 1d array of data"
  "pdf_data:dict of {label:prob density value}"
  plt.title(title)
  for label,pdf in pdf_data.items():
    plt.plot(X_data,pdf,label=label)
  plt.legend()
  plt.xlabel(r"$x$")
  plt.ylabel(r"$f(x)$")  

```

```python
#  normal
data=np.random.randn(200)
ax=plt.subplot()
sns.histplot(data,kde=True,ax=ax,stat="density",bins=20)
_=ax.set(title="Historgram of observed data",xlabel="x",ylabel="Freq.");
plt.tight_layout()
plt.show()
```

```python
#  comparing
x=np.linspace(0,8,100)
pdf_data={
  r"$\mu=4,\sigma=0.5$":sp.stats.norm.pdf(x,4,0.5),
  r"$\mu=4\sigma=1$":sp.stats.norm.pdf(x,4,1),
  r"$\mu=4,\sigma=2$":sp.stats.norm.pdf(x,4,2)
}
plot(x,pdf_data,title="Normal distribution with different parameters")
```

```python
# Gamma distribution
shape, scale = 2, 2
x=np.random.gamma(shape,scale,size=300)
ax=plt.subplot()
sns.histplot(x,kde=True,ax=ax,stat="density",bins=20)
_=ax.set(title="Historgram of observed data",xlabel="x",ylabel="Freq.");
plt.tight_layout()
plt.show()
None
```

```python
#  comapring gamma distrib
x=np.linspace(0,8,300)
pdf_data={
  r"$\alpha=1,\beta=0.5$":sp.stats.gamma.pdf(x,a=1,scale=2),
  r"$\alpha=2,\beta=0.5$":sp.stats.gamma.pdf(x,a=2,scale=2),
  r"$\alpha=3,\beta=0.5$":sp.stats.gamma.pdf(x,a=3,scale=2)
}
plot(x,pdf_data,title="Gamma distribution with different parameters")
```

```python
#  inverse gamma
# take same parameters as gamma
pdf_data={
  r"$\alpha=1,\beta=0.5$":sp.stats.invgamma.pdf(x,a=1,scale=2),
  r"$\alpha=2,\beta=0.5$":sp.stats.invgamma.pdf(x,a=2,scale=2),
  r"$\alpha=3,\beta=0.5$":sp.stats.invgamma.pdf(x,a=3,scale=2)
}
plot(x,pdf_data,title="Inverse Gamma distribution with different parameters")
```
